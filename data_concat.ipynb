```python
import os
import pandas as pd

# CSV 파일이 저장된 폴더 경로
folder_path = 'csv/'

# 폴더 내의 모든 파일 목록 가져오기
file_list = os.listdir(folder_path)

# 파일 이름을 기준으로 정렬
file_list.sort()

# CSV 파일만 필터링
csv_files = [file for file in file_list if file.endswith('.csv')]

print(csv_files)
df_list = []
for file in csv_files:
    file_path=folder_path + file
    print(file_path)
    df_list.append(pd.read_csv(file_path))
```

```python
df_list[1]
```

```python
df = pd.concat(df_list,ignore_index=True)
df
```

```python
df.to_csv('hyundai_used_car.csv')
```
